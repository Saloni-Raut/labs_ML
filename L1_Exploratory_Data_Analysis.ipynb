# **Handling Missing Data**

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.fillna(df.mean(), inplace=True)
```

```python
df
```

# **Data Handling**

```python
x = df['ph']
y = df['Hardness']
plt.scatter(x, y)
plt.show()
```

# **Data Visualisation**

```python
plt.figure(figsize=(10, 6))

sns.scatterplot(x='Hardness', y='Sulfate', hue='Potability', data=df, palette='Set1')
plt.title("Scatter Plot of Hardness vs Sulfate by Potability")
plt.show()

plt.figure(figsize=(10, 6))

sns.scatterplot(x='Conductivity', y='Organic_carbon', hue='Potability', data=df, palette='Set2')
plt.title("Scatter Plot of Conductivity vs Organic Carbon by Potability")
plt.show()
```

```python
plt.figure(figsize=(15, 8))
sns.boxplot(data=df)
plt.xticks(rotation=90)
plt.show()
```
